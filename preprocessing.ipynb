```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('data/all_df_2000_2010.csv')
```

```python
df['DATA'] = pd.to_datetime(df['DATA'])
df = df.set_index('DATA')
df = df.sort_index()
```

```python
# df = df['2018-01-01':'2018-01-30']
```

```python
df.dtypes
```

```python
df.head()
```

```python
df = df.astype(str)
df = df.apply(lambda x: x.str.replace(',','.'))
```

```python
df["PRECIPITACAO"]
```

```python
numeric_values = [
    'PRECIPITACAO',
    'P.A. NIVEL DA ESTACAO H',
    'P.A. MAX. H ANT',
    'P.A. MIN. NA HORA ANT',
    'RAD. GLOBAL',
    'TEMP AR - BULBO SECO',
    'TEMP PONTO DE ORVALHO',
    'TEMP MAX H ANT',
    'TEMP MIN H ANT',
    'TEMP ORVALHO MAX. H ANT',
    'TEMP ORVALHO MIN. H ANT',
    'UMIDADE REL. MAX. H ANT %',
    'UMIDADE REL. MIN. H ANT %',
    'UMIDADE RELATIVA DO AR HORARIA %',
    'VENTO DIR H',
    'VENTO RAJADA MAX',
    'VENTO VELOCIDADE H'
]
```

```python
list_astype = {x:"float64" for x in numeric_values}
```

```python
df = df.astype(list_astype, errors='ignore')
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df[numeric_values] = df[numeric_values].replace({-9999.0000: np.nan })
df[numeric_values] = df[numeric_values].replace({-9999.000: np.nan })
```

```python
df.describe()
```

```python
df_max = pd.DataFrame()
df_min = pd.DataFrame()
df_mean = pd.DataFrame()
df_final = pd.DataFrame()
```

```python
df_max = df.groupby(['DATA', 'estacao']).max()[['P.A. MAX. H ANT', 'TEMP MAX H ANT', 'TEMP ORVALHO MAX. H ANT', 'VENTO RAJADA MAX']]
```

```python
df_min = df.groupby(['DATA', 'estacao']).min()[['P.A. MIN. NA HORA ANT', 'TEMP MIN H ANT', 'UMIDADE REL. MIN. H ANT %']]
```

```python
df_mean = df.groupby(['DATA', 'estacao']).mean()[['PRECIPITACAO', 'P.A. NIVEL DA ESTACAO H', 'RAD. GLOBAL', 'TEMP AR - BULBO SECO', 'TEMP PONTO DE ORVALHO', 'UMIDADE RELATIVA DO AR HORARIA %', 'VENTO DIR H', 'VENTO VELOCIDADE H']]
```

```python
df_final = df_max.join(df_min).join(df_mean)
df_final.columns = [
    'pa_max_h_ant', 
    'temp_max_h_ant', 
    'temp_orvalho_max_h_ant', 
    'vento_rajada_max',
    'pa_min_na_hora_ant', 
    'temp_min_h_ant', 
    'umidade_rel_min_h_ant', 
    'precipitacao', 
    'pa_nivel_da_estacao_h',
    'rad_global', 
    'temp_ar_bulbo_seco',
    'temp_ponto_orvalho',
    'umidade_relativa_do_ar_horaria_percentage', 
    'vento_dir_h', 
    'velocidade_vento_h'
]
```
